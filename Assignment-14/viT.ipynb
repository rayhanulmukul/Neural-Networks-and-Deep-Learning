```python
# ===============================================================
# Auto-download 20-class ImageNet-derived dataset + ViT training
# (Imagenette 10 + ImageWoof 10)
# Works on ~8GB GPU (mixed precision + small ViT, 192x192)
# ===============================================================

import os, sys, tarfile, shutil, urllib.request, math, glob
from pathlib import Path
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

print("TensorFlow:", tf.__version__)
print("GPUs:", tf.config.list_physical_devices('GPU'))

# -------------------- Paths & URLs --------------------
DATA_ROOT = Path("./imagenet20_data")     # merged dataset output (train/val)
DOWNLOADS = Path("./downloads")           # where .tgz files will be stored
RAW_DIR   = Path("./raw")                 # where archives get extracted
IMAGENETTE_URL = "https://s3.amazonaws.com/fast-ai-imageclas/imagenette2-320.tgz"
IMAGEWOOF_URL   = "https://s3.amazonaws.com/fast-ai-imageclas/imagewoof2-320.tgz"

# -------------------- Repro & GPU ergonomics --------------------
SEED = 42
tf.random.set_seed(SEED); np.random.seed(SEED)

# Dont pre-allocate all VRAM
for g in tf.config.list_physical_devices('GPU'):
    try:
        tf.config.experimental.set_memory_growth(g, True)
    except Exception as e:
        print("Memory growth not set:", e)

# Mixed precision helps on Ampere+ GPUs (saves VRAM)
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy("mixed_float16")

# -------------------- Training config (8GB friendly) --------------------
IMG_SIZE     = 192
PATCH_SIZE   = 16
NUM_CLASSES  = 20
BATCH_SIZE   = 16     # if OOM -> 8
EPOCHS       = 30
AUTOTUNE     = tf.data.AUTOTUNE

# Small ViT
PROJECTION_DIM     = 128
NUM_HEADS          = 4      # must divide PROJECTION_DIM
TRANSFORMER_LAYERS = 6
MLP_UNITS          = [256, 128]
DROPOUT_RATE       = 0.1
MLP_DROPOUT        = 0.1
WEIGHT_DECAY       = 1e-4
BASE_LR            = 3e-4

# ===============================================================
# STEP 1  Download & Extract (Imagenette + ImageWoof)
# ===============================================================
DOWNLOADS.mkdir(parents=True, exist_ok=True)
RAW_DIR.mkdir(parents=True, exist_ok=True)
DATA_ROOT.mkdir(parents=True, exist_ok=True)

def download(url: str, out_path: Path):
    if out_path.exists():
        print(f"[skip] {out_path.name} already exists.")
        return
    print(f"[download] {url}")
    urllib.request.urlretrieve(url, out_path)
    print(f"[done] -> {out_path}")

def extract(tgz_path: Path, out_dir: Path):
    marker = out_dir / (tgz_path.stem + "_EXTRACTED")
    if marker.exists():
        print(f"[skip] already extracted -> {out_dir}")
        return
    print(f"[extract] {tgz_path} -> {out_dir}")
    with tarfile.open(tgz_path, "r:gz") as tar:
        tar.extractall(path=out_dir)
    marker.touch()
    print(f"[done] extracted")

# Download archives
imagenette_tgz = DOWNLOADS / "imagenette2-320.tgz"
imagewoof_tgz  = DOWNLOADS / "imagewoof2-320.tgz"
download(IMAGENETTE_URL, imagenette_tgz)
download(IMAGEWOOF_URL,  imagewoof_tgz)

# Extract to RAW_DIR
extract(imagenette_tgz, RAW_DIR)
extract(imagewoof_tgz,  RAW_DIR)

# Expect these top-level folders after extract:
# RAW_DIR/imagenette2-320/{train,val}/<10 classes>
# RAW_DIR/imagewoof2-320/{train,val}/<10 classes>
IMAGENETTE_DIR = RAW_DIR / "imagenette2-320"
IMAGEWOOF_DIR  = RAW_DIR / "imagewoof2-320"
assert IMAGENETTE_DIR.exists(), f"Missing {IMAGENETTE_DIR}"
assert IMAGEWOOF_DIR.exists(),  f"Missing {IMAGEWOOF_DIR}"

# ===============================================================
# STEP 2  Merge into a single 20-class dataset
# structure: DATA_ROOT/train/<20 classes>/..., DATA_ROOT/val/<20 classes>/...
# We prefix class folders to avoid any name collisions.
# ===============================================================
def copy_tree(src_root: Path, dst_root: Path, prefix: str):
    dst_root.mkdir(parents=True, exist_ok=True)
    for cls_dir in sorted([p for p in src_root.iterdir() if p.is_dir()]):
        cls_name = f"{prefix}_{cls_dir.name}"
        out_cls = dst_root / cls_name
        out_cls.mkdir(parents=True, exist_ok=True)
        # Copy all images in the class dir (non-recursive is enough for these datasets)
        for img_path in cls_dir.glob("*"):
            if img_path.is_file():
                shutil.copy2(img_path, out_cls / img_path.name)

# Clean any previous merge
for sub in ["train", "val"]:
    p = DATA_ROOT / sub
    if p.exists():
        shutil.rmtree(p)
    p.mkdir(parents=True, exist_ok=True)

# Merge train
copy_tree(IMAGENETTE_DIR / "train", DATA_ROOT / "train", prefix="nette")
copy_tree(IMAGEWOOF_DIR  / "train", DATA_ROOT / "train", prefix="woof")

# Merge val
copy_tree(IMAGENETTE_DIR / "val", DATA_ROOT / "val", prefix="nette")
copy_tree(IMAGEWOOF_DIR  / "val", DATA_ROOT / "val", prefix="woof")

# Quick sanity check
classes_train = sorted([d.name for d in (DATA_ROOT / "train").iterdir() if d.is_dir()])
classes_val   = sorted([d.name for d in (DATA_ROOT / "val").iterdir() if d.is_dir()])
print(f"Train classes: {len(classes_train)} | Val classes: {len(classes_val)}")
print("Sample classes:", classes_train[:5], "...", classes_train[-5:])

# ===============================================================
# STEP 3  tf.data pipelines from directory
# ===============================================================
def make_ds(root_dir: Path, subset: str):
    ds = tf.keras.utils.image_dataset_from_directory(
        root_dir,
        labels="inferred",
        label_mode="int",
        class_names=None,             # infer from folders
        color_mode="rgb",
        batch_size=BATCH_SIZE,
        image_size=(IMG_SIZE, IMG_SIZE),
        shuffle=(subset == "train"),
        seed=SEED
    )
    return ds

train_ds = make_ds(DATA_ROOT / "train", subset="train")
val_ds   = make_ds(DATA_ROOT / "val",   subset="val")

# Capture the inferred, stable class order (0..19)
class_names = train_ds.class_names
print("Final class order (0..19):", class_names)

# Normalize + augment
def norm(img, label):   return (tf.cast(img, tf.float32) / 255.0, label)
def aug(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, 0.2)
    img = tf.image.random_contrast(img, 0.8, 1.2)
    return img, label

train_ds = (train_ds.map(norm, num_parallel_calls=AUTOTUNE)
                     .map(aug,  num_parallel_calls=AUTOTUNE)
                     .prefetch(AUTOTUNE))
val_ds   = (val_ds.map(norm, num_parallel_calls=AUTOTUNE)
                 .prefetch(AUTOTUNE))

# ===============================================================
# STEP 4  ViT (small)
# ===============================================================
class ClassToken(layers.Layer):
    def __init__(self, dim, **kwargs):
        super().__init__(**kwargs); self.dim = dim
    def build(self, _):
        self.cls = self.add_weight("cls", shape=(1,1,self.dim),
                                   initializer="zeros", trainable=True)
    def call(self, x):
        b = tf.shape(x)[0]
        return tf.tile(self.cls, [b,1,1])

class AddPositionEmbs(layers.Layer):
    def __init__(self, n, dim, **kwargs):
        super().__init__(**kwargs); self.n = n; self.dim = dim
    def build(self, _):
        self.pe = self.add_weight("pos_emb",
                                  shape=(1, self.n, self.dim),
                                  initializer=keras.initializers.RandomNormal(stddev=0.02),
                                  trainable=True)
    def call(self, x):
        return x + self.pe

def mlp(x, units, drop):
    for u in units:
        x = layers.Dense(u, activation=tf.nn.gelu)(x)
        x = layers.Dropout(drop)(x)
    return x

def transformer_block(x, heads, dim, mlp_units, drop):
    y = layers.LayerNormalization(epsilon=1e-6)(x)
    y = layers.MultiHeadAttention(num_heads=heads, key_dim=dim//heads, dropout=drop)(y, y)
    x = layers.Add()([x, y])
    y = layers.LayerNormalization(epsilon=1e-6)(x)
    y = mlp(y, mlp_units, drop)
    return layers.Add()([x, y])

def build_vit(image_size=IMG_SIZE, patch_size=PATCH_SIZE, dim=PROJECTION_DIM,
              layers_n=TRANSFORMER_LAYERS, heads=NUM_HEADS, mlp_units=MLP_UNITS,
              num_classes=NUM_CLASSES, drop=DROPOUT_RATE, mlp_drop=MLP_DROPOUT):
    inp = keras.Input(shape=(image_size, image_size, 3))
    # Patchify + project with Conv2D
    x = layers.Conv2D(dim, patch_size, strides=patch_size, padding="valid", name="patch_proj")(inp)
    n = (image_size // patch_size) ** 2
    x = layers.Reshape((n, dim), name="patch_flatten")(x)
    # CLS token + positional embeddings
    cls = ClassToken(dim, name="cls_token")(x)
    x = layers.Concatenate(axis=1, name="prepend_cls")([cls, x])  # [B, 1+N, D]
    x = AddPositionEmbs(n+1, dim, name="pos_emb")(x)
    x = layers.Dropout(drop, name="emb_dropout")(x)
    # Transformer encoders
    for _ in range(layers_n):
        x = transformer_block(x, heads, dim, mlp_units, drop)
    # Head on CLS
    x = layers.LayerNormalization(epsilon=1e-6)(x)
    cls_tok = x[:, 0]
    x = layers.Dropout(mlp_drop)(cls_tok)
    x = layers.Dense(mlp_units[0], activation=tf.nn.gelu)(x)
    x = layers.Dropout(mlp_drop)(x)
    out = layers.Dense(num_classes, activation="softmax", dtype="float32")(x)  # keep float32 for stability
    return keras.Model(inp, out, name="ViT_Small_20cls")

model = build_vit()
model.summary()

# ===============================================================
# STEP 5  Compile & Train
# ===============================================================
# AdamW if available; else Adam
try:
    optimizer = tf.keras.optimizers.experimental.AdamW(learning_rate=BASE_LR, weight_decay=WEIGHT_DECAY)
except Exception:
    optimizer = tf.keras.optimizers.Adam(learning_rate=BASE_LR)
    print("Falling back to Adam (no weight decay).")

model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

ckpt_path = "vit20_best.keras"
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(ckpt_path, monitor="val_accuracy", save_best_only=True, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=6, restore_best_weights=True, verbose=1),
    tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1),
]

history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks)

# ===============================================================
# STEP 6  Training Curves
# ===============================================================
def plot_training(history):
    h = history.history
    ep = range(1, len(h["loss"]) + 1)

    plt.figure(figsize=(6,4))
    plt.plot(ep, h["loss"], label="Train Loss")
    plt.plot(ep, h["val_loss"], label="Val Loss")
    plt.xlabel("Epoch"); plt.ylabel("Loss")
    plt.title("ViT (Imagenette+ImageWoof 20)  Loss")
    plt.grid(True, alpha=0.3); plt.legend(); plt.show()

    plt.figure(figsize=(6,4))
    plt.plot(ep, h["accuracy"], label="Train Acc")
    plt.plot(ep, h["val_accuracy"], label="Val Acc")
    plt.xlabel("Epoch"); plt.ylabel("Accuracy")
    plt.title("ViT (Imagenette+ImageWoof 20)  Accuracy")
    plt.grid(True, alpha=0.3); plt.legend(); plt.show()

plot_training(history)

# ===============================================================
# STEP 7  Confusion Matrix + (optional) Report
# ===============================================================
val_loss, val_acc = model.evaluate(val_ds, verbose=1)
print(f"\nValidation accuracy: {val_acc:.4f}")

# Collect predictions
y_true, y_pred = [], []
for imgs, lbls in val_ds:
    probs = model.predict(imgs, verbose=0)
    y_true.extend(lbls.numpy().tolist())
    y_pred.extend(np.argmax(probs, axis=-1).tolist())
y_true = np.array(y_true, dtype=np.int32)
y_pred = np.array(y_pred, dtype=np.int32)

cm = tf.math.confusion_matrix(y_true, y_pred, num_classes=len(class_names)).numpy()
cm_norm = cm / (cm.sum(axis=1, keepdims=True) + 1e-12)

def plot_confusion(cm, labels, normalize=True, figsize=(11,10)):
    data = cm_norm if normalize else cm
    plt.figure(figsize=figsize)
    plt.imshow(data, interpolation="nearest")
    plt.title("Confusion Matrix" + (" (Normalized)" if normalize else ""))
    plt.colorbar()
    ticks = np.arange(len(labels))
    plt.xticks(ticks, labels, rotation=90)
    plt.yticks(ticks, labels)
    fmt = ".2f" if normalize else "d"
    thr = data.max()/2.
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            plt.text(j, i, format(data[i,j], fmt), ha="center",
                     color="white" if data[i,j] > thr else "black", fontsize=7)
    plt.ylabel("True"); plt.xlabel("Predicted")
    plt.tight_layout(); plt.show()

plot_confusion(cm, class_names, normalize=True)

try:
    from sklearn.metrics import classification_report
    print("\nClassification report:")
    print(classification_report(y_true, y_pred, target_names=class_names, digits=4))
except Exception:
    print("\n[Info] scikit-learn not installed; skipping classification report.")

# ===============================================================
# STEP 8  Sample Predictions Grid
# ===============================================================
def show_samples(ds, class_names, model, n=12):
    imgs, lbls = next(iter(ds.unbatch().batch(n)))
    probs = model.predict(imgs, verbose=0)
    preds = np.argmax(probs, axis=-1)

    rows, cols = 3, math.ceil(n/3)
    plt.figure(figsize=(14, 8))
    for i in range(n):
        ax = plt.subplot(rows, cols, i+1)
        ax.imshow(imgs[i].numpy())
        t = class_names[int(lbls[i])]
        p = class_names[int(preds[i])]
        ax.set_title(f"T:{t}\nP:{p} {'' if t==p else ''}", fontsize=9)
        ax.axis("off")
    plt.tight_layout(); plt.show()

show_samples(val_ds, class_names, model, n=12)

# ===============================================================
# STEP 9  Save model
# ===============================================================
model.save("vit20_final.keras")
print("Saved -> vit20_final.keras")
```

```python
# ============================================
# STEP 0  Imports
# ============================================
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# ============================================
# STEP 1  Load Dataset (ImageNet 20 subset)
# ============================================
#    CIFAR-10   (  ImageNet subset  )
(ds_train, ds_val), info = tfds.load(
    "cifar10",   # <--  "imagenet2012_subset"  1000 class,   20 class  
    split=["train", "test"],
    as_supervised=True,
    with_info=True
)

NUM_CLASSES = 20   #  20 classes   model 
IMG_SIZE = 64      #  
BATCH_SIZE = 64

def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.one_hot(label, NUM_CLASSES)
    return image, label

ds_train = ds_train.map(preprocess).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
ds_val   = ds_val.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# ============================================
# STEP 2  Define Models
# ============================================

# 2.1 Vision Transformer (ViT)
def build_vit_classifier():
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    # Patchify
    patches = layers.Conv2D(64, kernel_size=16, strides=16)(inputs)
    patches = layers.Reshape((-1, 64))(patches)

    # Transformer Encoder
    x = layers.LayerNormalization()(patches)
    x = layers.MultiHeadAttention(num_heads=4, key_dim=64)(x, x)
    x = layers.Add()([x, patches])
    x = layers.LayerNormalization()(x)
    x = layers.Dense(128, activation="gelu")(x)

    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    return keras.Model(inputs, outputs, name="ViT")

# 2.2 CNN Classifier
def build_cnn_classifier():
    model = keras.Sequential([
        layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
        layers.Conv2D(32, 3, activation='relu'),
        layers.MaxPooling2D(),
        layers.Conv2D(64, 3, activation='relu'),
        layers.MaxPooling2D(),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model

# 2.3 Fully Connected Feed-Forward NN
def build_fcfnn_classifier():
    model = keras.Sequential([
        layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(256, activation='relu'),
        layers.Dense(NUM_CLASSES, activation='softmax')
    ])
    return model

# ============================================
# STEP 3  Compile and Train Models
# ============================================
models = {
    "ViT": build_vit_classifier(),
    "CNN": build_cnn_classifier(),
    "FCFNN": build_fcfnn_classifier()
}

histories = {}
for name, model in models.items():
    print(f"\n Training {name} model...")
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    history = model.fit(ds_train, validation_data=ds_val, epochs=5)  # epoch   
    histories[name] = history

# ============================================
# STEP 4  Compare Performance
# ============================================
plt.figure(figsize=(12,6))
for name, history in histories.items():
    plt.plot(history.history["val_accuracy"], label=f"{name} Val Acc")
plt.legend()
plt.title("Validation Accuracy Comparison")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()

# ============================================
# STEP 5  Final Evaluation
# ============================================
for name, model in models.items():
    loss, acc = model.evaluate(ds_val)
    print(f"{name} Accuracy: {acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

# =====================================================
# STEP 1  Dataset Load (example: CIFAR-10 instead of ImageNet20 for demo)
# =====================================================
(ds_train, ds_val), info = tfds.load(
    "cifar10",
    split=["train", "test"],
    as_supervised=True,
    with_info=True
)

NUM_CLASSES = 10   # CIFAR-10 has 10 classes (for demo, you can replace with ImageNet20 subset)
IMG_SIZE = 64
BATCH_SIZE = 64

def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.one_hot(label, NUM_CLASSES)
    return image, label

ds_train = ds_train.map(preprocess).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
ds_val   = ds_val.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# =====================================================
# STEP 2  ViT Model Builder with variable num_heads
# =====================================================
def build_vit_classifier(num_heads=4):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    # Patchify
    patches = layers.Conv2D(64, kernel_size=16, strides=16)(inputs)
    patches = layers.Reshape((-1, 64))(patches)

    # Transformer Encoder with variable heads
    x = layers.LayerNormalization()(patches)
    x = layers.MultiHeadAttention(num_heads=num_heads, key_dim=64)(x, x)
    x = layers.Add()([x, patches])
    x = layers.LayerNormalization()(x)
    x = layers.Dense(128, activation="gelu")(x)

    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    return keras.Model(inputs, outputs, name=f"ViT_heads_{num_heads}")

# =====================================================
# STEP 3  Train models with different heads
# =====================================================
head_configs = [2, 4, 8]
histories = {}

for h in head_configs:
    print(f"\n Training ViT with num_heads={h}")
    model = build_vit_classifier(num_heads=h)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    history = model.fit(ds_train, validation_data=ds_val, epochs=5, verbose=1)
    histories[h] = history

# =====================================================
# STEP 4  Plot comparison
# =====================================================
plt.figure(figsize=(10,6))
for h, history in histories.items():
    plt.plot(history.history["val_accuracy"], label=f"heads={h}")
plt.title("Effect of Number of Heads on ViT Performance")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.show()

# =====================================================
# STEP 5  Final Accuracy Report
# =====================================================
for h, history in histories.items():
    final_acc = history.history["val_accuracy"][-1]
    print(f"ViT with {h} heads  Final Val Accuracy: {final_acc:.4f}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

# =====================================================
# STEP 1  Load CIFAR-10 (for demo, replace with ImageNet20 subset later)
# =====================================================
(ds_train, ds_val), info = tfds.load(
    "cifar10",
    split=["train", "test"],
    as_supervised=True,
    with_info=True
)

NUM_CLASSES = 10
IMG_SIZE = 64
BATCH_SIZE = 64

def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.one_hot(label, NUM_CLASSES)
    return image, label

ds_train = ds_train.map(preprocess).shuffle(1000).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
ds_val   = ds_val.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# =====================================================
# STEP 2  Different Patch Embedding Choices
# =====================================================

def linear_patch_embedding(inputs, patch_size=16, embed_dim=64):
    """Classic: flatten + Dense"""
    patches = tf.image.extract_patches(
        images=inputs,
        sizes=[1, patch_size, patch_size, 1],
        strides=[1, patch_size, patch_size, 1],
        rates=[1,1,1,1],
        padding="VALID"
    )
    patch_dims = patches.shape[-1]
    patches = layers.Reshape((-1, patch_dims))(patches)
    return layers.Dense(embed_dim)(patches)

def conv_patch_embedding(inputs, patch_size=16, embed_dim=64):
    """Conv2D as patchifier"""
    x = layers.Conv2D(embed_dim, kernel_size=patch_size, strides=patch_size)(inputs)
    x = layers.Reshape((-1, embed_dim))(x)
    return x

def hybrid_patch_embedding(inputs, embed_dim=64):
    """CNN backbone before patchification"""
    x = layers.Conv2D(32, 3, strides=2, activation="relu")(inputs)
    x = layers.Conv2D(64, 3, strides=2, activation="relu")(x)
    x = layers.Reshape((-1, 64))(x)
    return layers.Dense(embed_dim)(x)

# =====================================================
# STEP 3  ViT Block
# =====================================================
def transformer_block(x, num_heads=4, embed_dim=64, mlp_dim=128):
    # LayerNorm + MHA
    skip = x
    x = layers.LayerNormalization()(x)
    x = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(x, x)
    x = layers.Add()([x, skip])

    # MLP
    skip = x
    x = layers.LayerNormalization()(x)
    x = layers.Dense(mlp_dim, activation="gelu")(x)
    x = layers.Dense(embed_dim)(x)
    return layers.Add()([x, skip])

def build_vit(patch_type="linear", patch_size=16, embed_dim=64, num_heads=4):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    if patch_type == "linear":
        x = linear_patch_embedding(inputs, patch_size, embed_dim)
    elif patch_type == "conv":
        x = conv_patch_embedding(inputs, patch_size, embed_dim)
    elif patch_type == "hybrid":
        x = hybrid_patch_embedding(inputs, embed_dim)
    else:
        raise ValueError("Unknown patch_type")

    # Add a few transformer blocks
    for _ in range(2):
        x = transformer_block(x, num_heads=num_heads, embed_dim=embed_dim)

    # Classification head
    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    return keras.Model(inputs, outputs, name=f"ViT_{patch_type}")

# =====================================================
# STEP 4  Train different patch embeddings
# =====================================================
patch_types = ["linear", "conv", "hybrid"]
histories = {}

for pt in patch_types:
    print(f"\n Training ViT with {pt} patch embedding")
    model = build_vit(patch_type=pt)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    history = model.fit(ds_train, validation_data=ds_val, epochs=5, verbose=1)
    histories[pt] = history

# =====================================================
# STEP 5  Plot comparison
# =====================================================
plt.figure(figsize=(10,6))
for pt, history in histories.items():
    plt.plot(history.history["val_accuracy"], label=f"{pt} embedding")
plt.title("Effect of Patch Embedding Choice on ViT Performance")
plt.xlabel("Epochs")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.show()

# =====================================================
# STEP 6  Final Report
# =====================================================
for pt, history in histories.items():
    acc = history.history["val_accuracy"][-1]
    print(f"Patch type = {pt}, Final Validation Accuracy = {acc:.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# ===============================
# CONFIG
# ===============================
IMG_SIZE   = 64
PATCH_SIZE = 16
EMBED_DIM  = 64
NUM_HEADS  = 4
NUM_CLASSES = 20
EPOCHS = 3
BATCH_SIZE = 64

# ===============================
# 1. Dummy Dataset (20 classes)
# ===============================
def get_dataset(num_samples=5000):
    x = np.random.rand(num_samples, IMG_SIZE, IMG_SIZE, 3).astype("float32")
    y = np.random.randint(0, NUM_CLASSES, size=(num_samples,))
    y = keras.utils.to_categorical(y, NUM_CLASSES)
    return x, y

x_train, y_train = get_dataset(8000)
x_val, y_val     = get_dataset(2000)

ds_train = tf.data.Dataset.from_tensor_slices((x_train, y_train)).shuffle(1024).batch(BATCH_SIZE)
ds_val   = tf.data.Dataset.from_tensor_slices((x_val, y_val)).batch(BATCH_SIZE)

print(" Dataset ready:", x_train.shape, y_train.shape)

# ===============================
# 2. Patch Embedding
# ===============================
def linear_patch_embedding(x, patch_size=PATCH_SIZE, embed_dim=EMBED_DIM):
    patches = layers.Conv2D(embed_dim, kernel_size=patch_size, strides=patch_size)(x)
    seq = tf.reshape(patches, (-1, (IMG_SIZE // patch_size) ** 2, embed_dim))
    return seq

# ===============================
# 3. Positional Embedding Layers
# ===============================
class LearnablePositionalEmbedding(layers.Layer):
    def __init__(self, seq_len, embed_dim):
        super().__init__()
        self.pos_emb = self.add_weight("pos_emb", shape=(1, seq_len, embed_dim))

    def call(self, x):
        return x + self.pos_emb

class SinusoidalPositionalEmbedding(layers.Layer):
    def __init__(self, seq_len, embed_dim):
        super().__init__()
        position = np.arange(seq_len)[:, np.newaxis]
        div_term = np.exp(np.arange(0, embed_dim, 2) * -(np.log(10000.0) / embed_dim))
        pe = np.zeros((seq_len, embed_dim))
        pe[:, 0::2] = np.sin(position * div_term)
        pe[:, 1::2] = np.cos(position * div_term)
        self.pe = pe[np.newaxis, ...]

    def call(self, x):
        # Cast dtype to match x (fixes float16/float32 mismatch)
        pe = tf.cast(self.pe, x.dtype)
        return x + pe

# ===============================
# 4. Transformer Block
# ===============================
def transformer_block(x, num_heads=NUM_HEADS, embed_dim=EMBED_DIM, ff_dim=128):
    attn_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(x, x)
    x = layers.Add()([x, attn_output])
    x = layers.LayerNormalization()(x)

    ff = keras.Sequential([
        layers.Dense(ff_dim, activation="relu"),
        layers.Dense(embed_dim),
    ])
    x = layers.Add()([x, ff(x)])
    x = layers.LayerNormalization()(x)
    return x

# ===============================
# 5. ViT Builder
# ===============================
def build_vit(pos_type="learnable", patch_size=PATCH_SIZE, embed_dim=EMBED_DIM, num_heads=NUM_HEADS):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = linear_patch_embedding(inputs, patch_size, embed_dim)
    seq_len = (IMG_SIZE // patch_size) ** 2

    if pos_type == "learnable":
        x = LearnablePositionalEmbedding(seq_len, embed_dim)(x)
    elif pos_type == "sinusoidal":
        x = SinusoidalPositionalEmbedding(seq_len, embed_dim)(x)
    elif pos_type == "none":
        pass

    for _ in range(2):
        x = transformer_block(x, num_heads=num_heads, embed_dim=embed_dim)

    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    return keras.Model(inputs, outputs)

# ===============================
# 6. Train & Compare
# ===============================
histories = {}
pos_types = ["learnable", "sinusoidal", "none"]

for pt in pos_types:
    print(f"\n Training ViT with {pt} positional embedding")
    model = build_vit(pos_type=pt)
    model.compile(
        optimizer=keras.optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    history = model.fit(ds_train, validation_data=ds_val, epochs=EPOCHS, verbose=1)
    histories[pt] = history

# ===============================
# 7. Plot Results
# ===============================
plt.figure(figsize=(10,5))
for pt,h in histories.items():
    plt.plot(h.history['val_accuracy'], label=f"{pt} val_acc")
plt.title("Effect of Positional Embedding on ViT")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.legend()
plt.show()
```
